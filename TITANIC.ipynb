```python
#Kullanacağımız kütüphaneleri import ediyoruz.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Localimizdeki .csv dosyasını import ediyoruz.
df = pd.read_csv("asset-v1_Kodlasam+MS01+2023_06+type@asset+block@nobel.csv") 
df
```

```python
#Doğum tarihi sütununu 'datetime' formatına çevirip sadece yılını aldık, ödül aldıkları yıl ile farkını hesaplayıp yeni bir 'Age' sütunu oluşturduk.
df['birth_date'] = pd.to_datetime(df['birth_date'], format='%Y-%m-%d')
df['age'] = (df['year'] - df['birth_date'].dt.year)
df  
```

```python
#Yaş Dağılım Histogramı
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=10, kde=True)
plt.title('Yaş Dağılımı Histogramı')
plt.xlabel('Yaş')
plt.ylabel('Frekans')
plt.show();
```

```python
median_age = df['age'].median()
print(f"Medyan: {median_age}")
```

```python
mod_age = df['age'].mode()[0]
print(f"Mod : {mod_age}")
```

```python
median_age = df['age'].median()
df['age'].fillna(median_age, inplace=True)
#MOD,MEDYAN,ORTALAMA VE YAŞ DAĞILIMI HİSTOGRAMI İNCELENDİĞİNDE 'AGE' SÜTUNUNDA 
#BOŞ OLAN 28 BOŞ DEĞERİ MEDYAN İLE DOLDURDUĞUMUZDA VERİ SETİMİZ BÜYÜK BİR DEĞİŞİKLİĞE UĞRAMAZ 
```

```python
df['age'] = df['age'].astype(int)
df.info()
#Age sütunumuzu 'integer' tipine çevirdik.
```

```python
df['sex'].fillna('Bilinmiyor', inplace=True)
#Cinsiyet sütunumuzda 26 adet boş satır olduğunu görüyoruz, bu verilerin rastgele veya ağırlıkta olan 'erkek' cinsiyeti ile değiştirilmesi bazı soruların 
#yanlış sonuçlanmasına neden olacağı için 'bilinmiyor' şeklinde doldurduk.
```

```python
df['motivation'].fillna(df['category'], inplace=True)
#'motivation' sütunu altında boş olan satırlar yerine hangi kategoride çalışma yaptıysa o başlık ile doldurduk.
```

```python
df['birth_country'].fillna('Bilinmiyor', inplace=True)
#'birth_country' sütunu altında 26 adet boş satır olduğunu görüyoruz, boş satırları 'bilinmiyor' şeklinde dolduruyoruz.
```

```python
# Veri Setimizi incelediğimizde genelde 'organization_country' ile 'death_country' satırlarının aynı olduğunu görüyoruz.
#Boş olan 'organization_country' yerine 'death_country' satırları ile dolduruyoruz.
df['organization_country'] = df['organization_country'].fillna(df['death_country'])

# Sonuçları görüntüleme
df
```

```python
#Toplam boş satırlara bakıyoruz.
df.isnull().sum() 
```

# 4- Nobel ödülünü en çok kazanan ülkenin hangi yıldan itibaren hakimiyet sağladığını görselleştirip bu hakimiyette rol oynayan şeyler nelerdir? İçgörülerinizi paylaşır mısınız?

```python
# 'Bilinmiyor' olarak etiketlenmiş satırları filtreleyerek dahil etmiyoruz
nobel_filtered = df[df['organization_country'] != 'Bilinmiyor']

# Yıl ve ülke bazında ödül sayılarını hesaplıyoruz
country_year_counts = nobel_filtered.groupby(['year', 'organization_country']).size().unstack(fill_value=0)

# Yıllar boyunca en çok ödül kazanan ülkeyi buluyoruz
top_country_per_year = country_year_counts.idxmax(axis=1)
top_country_awards_per_year = country_year_counts.max(axis=1)

# Grafik için yıllar ve ülkeler listesini alıyoruz
years = top_country_per_year.index
countries = top_country_per_year.values

# Benzersiz ülkeleri almak için
unique_countries = top_country_per_year.unique()

# Her ülke için farklı renk ataması yapıyoruz
color_map = {country: idx for idx, country in enumerate(unique_countries)}
colors = [color_map[country] for country in countries]

# Grafik oluşturma
plt.figure(figsize=(16, 8))
scatter = plt.scatter(years, countries, c=colors, cmap='tab20b')

# Renk çubuğu ekleyin
cbar = plt.colorbar(scatter, ticks=range(len(unique_countries)))
cbar.ax.set_yticklabels(unique_countries)

# Grafiği etiketleyin ve gösterin
plt.title('Yıllara Göre En Çok Nobel Ödülü Kazanan Ülkeler', fontsize=16)
plt.xlabel('Yıl', fontsize=14)
plt.ylabel('Ülkeler', fontsize=14)
plt.grid(True)
plt.xticks(np.arange(min(years), max(years) + 1, 5), rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)  # Y eksenindeki etiket boyutunu ayarla
plt.gca().spines['top'].set_visible(False)  # Üst kenarı gizliyoruz
plt.gca().spines['right'].set_visible(False)  # Sağ kenarı gizliyoruz
plt.gca().set_facecolor('#f5f5dc')  # Arka plan rengini düzenliyoruz
plt.show()
```

# 5- Nobel Ödülü kazananların cinsiyetlerini, yaşlarını, ödül kategorisi ve yılları kullanarak görselleştiriniz.(Her bir ödül kategorisi için ayrı grafik gösteriniz) Çıkan sonuçlara göre görseli yorumlayınız.

```python
# 'Bilinmiyor' olarak etiketlenmiş satırları filtreleme
df_filtered = df[df['sex'] != 'Bilinmiyor']

# Ödül kategorilerine göre veriyi gruplama
grouped = df_filtered.groupby('category')

# Her bir ödül kategorisi için ayrı grafik oluşturma
for category, group in grouped:
    plt.figure(figsize=(20, 6))
    sns.scatterplot(data=group, x='year', y='age', hue='sex', palette='deep', s=100)
    plt.title(f'Nobel Ödülü Kazananların Yaş ve Cinsiyetleri - {category}')
    plt.xlabel('Yıl')
    plt.ylabel('Yaş')
    plt.legend(title='Cinsiyet')
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.show()
```

# 6-) 1938-1945 yılı arasında Nobel Ödülü kazananların kategorilerini ve ülkelerini görselleştirip yorumlayınız.


```python

# 'Bilinmiyor' olarak etiketlenmiş satırları filtreleyerek dahil etmiyoruz
df_filtered = df[df['organization_country'] != 'Bilinmiyor']

# 1938-1945 yılları arasındaki verileri filtreliyoruz
df_range = df_filtered[(df_filtered['year'] >= 1938) & (df_filtered['year'] <= 1945)]

# Ülkelere göre ödül sayılarını görselleştiriyoruz
plt.figure(figsize=(20, 8))
sns.countplot(data=df_range, x='organization_country', palette='viridis', order=df_range['organization_country'].value_counts().index)
plt.title('1938-1945 Yılları Arasında Nobel Ödülü Kazanan Ülkeler', fontsize=16)
plt.xlabel('Ülke', fontsize=14)
plt.ylabel('Ödül Sayısı', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()

# Ülkelere ve kategorilere göre ödül sayısını hesaplayalıyoruz
sum_df = df_range.groupby(['organization_country', 'category']).size().unstack(fill_value=0)

# Grafik oluşturuyoruz
plt.figure(figsize=(16, 6))
sum_df.plot(kind='bar', stacked=True, colormap='viridis', figsize=(20, 8))

plt.title('1938-1945 Yılları Arasında Nobel Ödülü Kazanan Kategoriler ve Ülkeler', fontsize=16)
plt.xlabel('Ülke', fontsize=14)
plt.ylabel('Toplam Ödül Sayısı', fontsize=14)
plt.xticks(rotation=45, ha='right')

plt.legend(title='Kategori', bbox_to_anchor=(1.05, 1), loc='upper left', shadow=True)
plt.tight_layout()

plt.show();
```

# 7. 1947-1991 yılları arasında Nobel Ödülü kazananların kategorilerini ve ülkelerini görselleştirip yorumlayınız.(Her kategori için ayrı bir grafik olması istenmektedir)

```python
# 1947-1991 arası yılları filtreleme
range2_df = filtered_df[(filtered_df['year'] >= 1947) & (filtered_df['year'] <= 1991)]
filtered_df = df[(df['organization_country'] != 'Bilinmiyor')]

# Her kategori için ayrı grafik çiziyoruz
categories = range2_df['category'].unique()
for category in categories:
    # Belirli bir kategoriye ait verileri alıyoruz
    category_df = range2_df[range2_df['category'] == category]
    
    # Ülke bazında ödül sayılarını hesaplıyoruz
    country_counts = category_df['organization_country'].value_counts()
    countries = country_counts.index.tolist()  # Ülkelerin isimleri
    counts = country_counts.tolist()  # Ödül sayıları
    
    # Seaborn renk paletlerinden birini kullanalım
    colors = sns.color_palette("rocket_r", len(countries))
    
    plt.figure(figsize=(20, 8))
    bars = plt.bar(countries, counts, color=colors)
    
    # Her barın üzerine ödül sayısını yazdırma
    for bar, count in zip(bars, counts):
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + 0.15, count, ha='center', va='bottom', fontsize=8)
    
    # Grafik başlığı ve eksen isimleri
    plt.title(f'1947-1991 Yılları Arasında {category} Nobel Ödülü Kazanan Ülkeler', fontsize=16)
    plt.xlabel('Ülke', fontsize=14)
    plt.ylabel('Ödül Sayısı', fontsize=14)
    plt.xticks(rotation=45, ha='right', fontsize=10)
    plt.yticks(fontsize=10)
    plt.grid(axis='y', linestyle='--', linewidth=0.5)
    
    plt.show()
```

# 7-) Kimya, Edebiyat, Barış, Fizik ve Tıp kategorilerindeki 2000 sonrasındaki kişilerin ülkelerini, yaşlarını görselleştirin.(Her bir Kategori için ayrı görselleştirme yapılması istenmektedir) Veriyi yorumlayınız.

```python
# Verileri 2000 sonrası yıllar arasında filtreleme 
range3_df = df[(df['year'] >= 2000) & (df['organization_country'] != 'Bilinmiyor')]

# Kategoriler listesi
categories = ['Chemistry', 'Literature', 'Peace', 'Physics', 'Medicine']

# Her kategori için ayrı görselleştirme yapmak
for category in categories:
    # Belirli bir kategoriye ait verileri al
    category_df = range3_df[range3_df['category'] == category]
    
    # Görselleştirme
    plt.figure(figsize=(20, 6))
    sns.scatterplot(data=category_df, x='organization_country', y='age', hue='organization_country', palette='viridis', legend=None, s=100)
    
    # Grafik başlığı ve eksen etiketleri
    plt.title(f'2000 Sonrası Nobel Ödülü Kazananların Yaş Dağılımı - {category}', fontsize=16)
    plt.xlabel('Ülke', fontsize=14)
    plt.ylabel('Yaş', fontsize=14)
    plt.xticks(rotation=45, ha='right', fontsize=12)
    plt.yticks(fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

```python

# Verileri 2000 sonrası yıllar arasında filtreliyoruz
range3_df = df[(df['year'] >= 2000) & (df['organization_country'] != 'Bilinmiyor')]

# Kategoriler listesi
categories = ['Chemistry', 'Literature', 'Peace', 'Physics', 'Medicine']

# Kategorilere göre yaş ortalamalarını hesaplıyoruz
category_averages = range3_df.groupby('category')['age'].mean().loc[categories]

# Grafik boyutları ve stilini ayarlıyoruz
plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")

# Renk paletini belirliyoru<
colors = sns.color_palette("viridis", len(categories))

# Çubuk grafiği oluşturuyoruz
barplot = sns.barplot(x=category_averages.index, y=category_averages.values, palette=colors)

# Başlık ve eksen etiketleri
plt.title('2000 Sonrası Nobel Ödülü Kazananların Kategori Bazında Ortalama Yaşları', fontsize=20, fontweight='bold')
plt.xlabel('Kategori', fontsize=16, fontweight='bold')
plt.ylabel('Ortalama Yaş', fontsize=16, fontweight='bold')
plt.xticks(fontsize=14, rotation=45, ha='right')
plt.yticks(fontsize=14)

# Çubukların üzerine yaş ortalamalarını yazdırıyoruz
for index, value in enumerate(category_averages.values):
    barplot.text(index, value + 0.5, f'{value:.1f}', color='black', ha="center", fontsize=12, fontweight='bold')

# Izgara çizgilerini ekliyoruz
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Grafik düzenlemeleri
plt.tight_layout()
plt.show()
```
